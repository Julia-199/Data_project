**2. Выявляем причины плохих оценок по качеству еды**

Производим отбор только негативных оценок (1 и 2), 
проводим анализ - при каких условиях посетители оставляли больше всего негативных отзывов (количество, % соотношение от всех негативных отзывов)
проводим отбор "топ 20" комбинаций

Большее количество негативных отзывов было получено при комбинации:
Бизнес ланч в ресторане (не на вынос) у тех, кто предпочитает итальянскую кухню. 




```python
low_food = data[(data["FoodRating"] < 3)].groupby(["DiningOccasion", "MealType", "TimeOfVisit", "PreferredCuisine"]).count()
low_food = low_food.iloc[:, 0:1] 
low_food.rename(columns = {"CustomerID" : "quantity"}, inplace = True)
low_food["percent"] = low_food["quantity"] / low_food["quantity"].sum()
top_20 = low_food.nlargest(20, "quantity")
print(top_20)
print()
top_20.groupby(["DiningOccasion"])["quantity"].sum()
```

**3. при каких целях посещениях большинство негативных оценок**

Также у посетителей, посетивших ресторан с целью бизнеса в целом больше всего негативных оценок (81), немного меньше на праздничные мероприятия (74), значительно меньше при обычных визитах (49), это также отображено на графике


```python
sns.barplot(data = top_20, x = "DiningOccasion", y = "quantity", estimator = np.sum)
plt.title("Распределение негативных отзывов по цели посещения")
plt.show()
```

**4.насколько велика доля таких клиентов (важность критерия)**

Доля посетителей с целью "бизнес", "праздничные события" - 68% от общего количества гостей

```python
occation = data.groupby("DiningOccasion").count().reset_index()
#occation = occation.iloc[:, 0:2]

occation1 = occation['DiningOccasion']
occation2 = occation['CustomerID']
colors = sns.color_palette("pastel")[0:3]
plt.pie(occation2, labels = occation1, colors = colors, autopct="%.0f%%")
plt.title("Распределение общего кол-ва посетителей по цели посещения ресторана")
plt.show()
```

**8.Какие факторы коррелируют с со средней суммой чека**

Средний уровень чека выше во время ланча;
так же есть небольшая корреляция размера чека и кол-ва группы

```python
time_2 = data.groupby(["TimeOfVisit", "FoodRating", "AverageSpend"]).count().reset_index()
time_2 = time_2.iloc[:, 0:4]
time_2.rename(columns = {"CustomerID" : "Count"}, inplace = True)
time_2 = time_2.groupby(["TimeOfVisit"]).agg({"FoodRating" : "mean", "AverageSpend" : "mean", "Count" : "sum"})
time_2["persent_of_visits"] = time_2["Count"] / time_2["Count"].sum()
time_2 = time_2.drop("Count", axis = 1)
print(time_2)
```

**9. Демографический состав.**

Пропорции по возрасту клиентов: большая часть клиентов в возрастном диапазоне от 41 до 55 и от 56 до 69

```python
age55 = len(data[(data["Age"] > 55)])
age40 = len(data[(data["Age"] > 40) & (data["Age"] <= 55)])
age26 = len(data[(data["Age"] > 26) & (data["Age"] <= 40)])
age18 = len(data[(data["Age"] > 18) & (data["Age"] <= 26)])

age = [age18, age26, age40, age55]
age_label = ["18-26", "27-40", "41-55", "56-69"]


colors = sns.color_palette("bright")[0:4]
plt.pie(age, labels = age_label, colors = colors, autopct="%.0f%%")
plt.title("Распределение общего кол-ва посетителей по возрасту")
plt.show()

```
